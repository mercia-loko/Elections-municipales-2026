## Étape 4 — Visualisations
### 📊 Graphique 1 : Parité globale femmes / hommes

````python
parite = df['SEXPSN'].value_counts()
labels = ['Femmes' if x == 'F' else 'Hommes' for x in parite.index]
colors = ['#e74c3c', '#2980b9']

fig, ax = plt.subplots(figsize=(7, 7))
wedges, texts, autotexts = ax.pie(
    parite, labels=labels, autopct='%1.1f%%',
    colors=colors, startangle=90,
    textprops={'fontsize': 13}
)
for at in autotexts:
    at.set_fontsize(14)
    at.set_fontweight('bold')
    at.set_color('white')

ax.set_title('Parité femmes / hommes parmi les élus\n(1er tour — Paris, Lyon, Marseille — 2026)',
             fontsize=14, fontweight='bold', pad=20)
plt.tight_layout()
plt.savefig('graphique1_parite_globale.png', dpi=150, bbox_inches='tight')
plt.show()
print(f"Femmes : {parite.get('F', 0)} | Hommes : {parite.get('M', 0)}")
````

### 📊 Graphique 2 : Parité par ville

````python
parite_ville = df.groupby(['VILLE', 'SEXPSN']).size().unstack(fill_value=0)
parite_ville.columns = ['Femmes' if c == 'F' else 'Hommes' for c in parite_ville.columns]
parite_pct = parite_ville.div(parite_ville.sum(axis=1), axis=0) * 100

fig, ax = plt.subplots(figsize=(9, 5))
parite_pct.plot(kind='bar', ax=ax, color=['#e74c3c', '#2980b9'],
                width=0.5, edgecolor='white')

ax.axhline(50, color='black', linestyle='--', linewidth=1, label='Seuil parité 50%')
ax.set_ylabel('Pourcentage (%)')
ax.set_xlabel('')
ax.set_title('Parité femmes / hommes par ville (% des élus)', fontsize=13, fontweight='bold')
ax.set_xticklabels(parite_pct.index, rotation=0)
ax.legend()
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
plt.tight_layout()
plt.savefig('graphique2_parite_par_ville.png', dpi=150, bbox_inches='tight')
plt.show()
````

### 📊 Graphique 3 : Nuances politiques par ville

````python
politique_ville = df.groupby(['VILLE', 'NUANCE_LABEL']).size().unstack(fill_value=0)

fig, ax = plt.subplots(figsize=(13, 6))
politique_ville.plot(kind='bar', ax=ax, width=0.7, edgecolor='white')
ax.set_ylabel('Nombre d\'élus')
ax.set_xlabel('')
ax.set_title('Répartition des nuances politiques par ville\n(élus au 1er tour 2026)',
             fontsize=13, fontweight='bold')
ax.set_xticklabels(politique_ville.index, rotation=0)
ax.legend(title='Nuance politique', bbox_to_anchor=(1.01, 1), loc='upper left', fontsize=9)
plt.tight_layout()
plt.savefig('graphique3_politique_par_ville.png', dpi=150, bbox_inches='tight')
plt.show()
````

### 📊 Graphique 4 : Distribution des âges des élus

````python
fig, ax = plt.subplots(figsize=(11, 5))
for ville, color in zip(['Paris', 'Lyon'], ['#8e44ad', '#27ae60']):
    subset = df[df['VILLE'] == ville]['AGE']
    ax.hist(subset, bins=12, alpha=0.6, color=color, label=ville, edgecolor='white')

ax.axvline(df['AGE'].mean(), color='black', linestyle='--',
           label=f"Âge moyen global : {df['AGE'].mean():.0f} ans")
ax.set_xlabel('Âge')
ax.set_ylabel('Nombre d\'élus')
ax.set_title('Distribution des âges des élus au 1er tour 2026', fontsize=13, fontweight='bold')
ax.legend()
plt.tight_layout()
plt.savefig('graphique4_ages.png', dpi=150, bbox_inches='tight')
plt.show()
````

## Étape 5 — Synthèse des insights

````python
nb_femmes = (df['SEXPSN'] == 'F').sum()
nb_hommes = (df['SEXPSN'] == 'M').sum()
pct_femmes = nb_femmes / len(df) * 100
age_moyen = df['AGE'].mean()
age_min = df['AGE'].min()
age_max = df['AGE'].max()
elu_plus_jeune = df.loc[df['AGE'].idxmin(), ['PREPSN', 'NOMPSN', 'AGE', 'VILLE']]
nuance_dominante = df['NUANCE_LABEL'].value_counts().idxmax()

print('=' * 55)
print('         SYNTHÈSE DES RÉSULTATS')
print('=' * 55)
print(f'👥 Total élus analysés     : {len(df)}')
print()
print(f'⚖️  Parité globale          : {pct_femmes:.1f}% de femmes')
if pct_femmes >= 50:
    print('   → Parité atteinte ✅')
else:
    print(f'   → Écart de {50 - pct_femmes:.1f} points par rapport à la parité ⚠️')
print()
print(f'🎂 Âge moyen               : {age_moyen:.0f} ans')
print(f'   Tranche : {age_min} – {age_max} ans')
print(f'   Élu(e) le/la plus jeune : {elu_plus_jeune["PREPSN"]} {elu_plus_jeune["NOMPSN"]}'
      f' ({int(elu_plus_jeune["AGE"])} ans, {elu_plus_jeune["VILLE"]})')
print()
print(f'🏆 Nuance dominante        : {nuance_dominante}')
print(f'   ({df["NUANCE_LABEL"].value_counts().iloc[0]} élus)')
print('=' * 55)
print()
print('💡 CONCLUSIONS')
print('-' * 55)
print('1. La parité est un enjeu visible dans ces élections :')
print('   les résultats varient significativement par ville.')
print('2. La classe politique locale reste dominée par des')
print('   profils de 40-60 ans, avec peu de jeunes élus.')
print('3. Les nuances politiques révèlent des dynamiques')
print('   locales très différentes entre Paris et Lyon.')
````
